```python
import pandas as pd
from sqlalchemy import create_engine,text
from sqlalchemy.engine.url import URL
```

```python
usuario = 'root'
senha = 'secret'

engine = create_engine(URL(
    drivername = 'mysql',
    username = usuario,
    password = senha,
    
    host = 'localhost',
    query = {'charset': 'utf8'},
    port = '3306', 
    database='classicmodels'
))

conn = engine.connect()
```

```python
pd.read_sql('show tables', con = conn)
```

```python
pd.read_sql('select * from customers limit 10', con = conn)
```

```python
pd.read_sql('select count(customerNumber), country from customers group by country order by count(customerNumber) desc', con = conn)
```

```python
pd.read_sql('select country from customers', con = conn).value_counts()
```

```python
# Definir a query SQL
query = 'SELECT * FROM products'

# Executar a query
products = pd.read_sql(query, conn)
```

```python
products
```

```python
productlines = pd.read_sql('select * from productlines', con = conn)
productlines
```

```python
pd.read_sql("select count(b.productCode), a.productLine, a.textDescription from productlines a "
            "inner join products b on a.productLine = b.productLine "
            "group by a.productLine", con = conn)
```

```python
productlines[['productLine', 'textDescription']].merge(products[['productCode','productLine']], 
                                                       how='inner', on='productLine'
                                                      )[['productLine', 'productCode']].groupby('productLine').count()
```

```python
productlines[['productLine', 'textDescription']].merge(products[['productCode','productLine']], 
                                                       how='inner', on='productLine'
                                                      )[['productLine', 'productCode']].groupby('productLine').count().merge(
                                            productlines[['productLine', 'textDescription']], 
                                            how='inner', on='productLine')
```

```python
pd.DataFrame(productlines[['productLine', 'textDescription']].merge(products[['productCode','productLine']], 
                                                       how='inner', on='productLine'
                                                      ).groupby(['productLine', 'textDescription'])['productLine'].count())
```
